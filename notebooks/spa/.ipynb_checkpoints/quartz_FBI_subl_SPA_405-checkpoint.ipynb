## Import silica spectra, normalize and subtract silica bg

```python
silica_file = '/Users/pabloherrero/sabat/laserlab/spectra/emission_silica_FBI_FBI_Ba_powder_exc_250nm.txt'
si = pd.read_csv(silica_file, skiprows=1, sep='\s', decimal='.', names=['fbi', 'fbiba', 'silica'], engine='python')

# si['fbi'] = si['fbi'].subtract(si['silica'])
si['fbiba'] = si['fbiba'].subtract(si['silica'])

norm = np.trapz(si['fbi'].values, x=si.index)
si /= norm
```

```python
si
```

## Interpolate to integer wavelengths

```python
from scipy.interpolate import interp1d

x = si.index.values
yf = si.fbi.values
yb = si.fbiba.values
ys = si.silica.values

f = interp1d(x, yf, kind='cubic')
f2 = interp1d(x, yb, kind='cubic')
f3 = interp1d(x, ys, kind='cubic')

xnew = np.arange(300, 794, 1)

plt.plot(x, yf, 'o', x, yb, 'o', xnew, f(xnew), '-', xnew, f2(xnew), '--')
plt.legend(['fbi', 'fbi-ba', 'interp', 'interp'], loc='best')
plt.show()
```

```python
siip = pd.DataFrame([xnew, f(xnew), f2(xnew), f3(xnew)]).T
siip.columns = ['wavelength', 'fbi', 'fbiba', 'silica']
siip.set_index('wavelength', inplace=True, drop=True)
```

```python
plt.plot(siip.index, siip.fbi, '-g', label='FBI')
plt.plot(siip.index, siip.fbiba, '-b', label='FBI-Ba')
plt.legend()
plt.gca().set(xlabel='Wavelength [nm]', ylabel='Rel intensity')
```

## Preview average images with axis projection

```python
fig, ax = plt.subplots(2,2, figsize=(16,16))

for i, f in enumerate(data):
    j, k = i//2, i%2
    d = io.imread(f).astype(float)
    bar = ax[j,k].imshow(d)
    plt.colorbar(bar, ax=ax[j,k])
    ax[j,k].set(title=names[i])
```

```python
from scipy.ndimage.interpolation import zoom
def projection_axis_bar(S:np.ndarray, rebin: int = 10, title: str = None):
    """Plot image and bar chart with projection on each axes
    S : image array
    rebin: step in the barcharts"""
    
    grid = plt.GridSpec(4,4, hspace=0.3, wspace=0.3)
    fig = plt.figure(figsize=(10,10))
    main_ax = fig.add_subplot(grid[:-1,1:])
    y_hist = fig.add_subplot(grid[:-1,0], sharey=main_ax )
    x_hist = fig.add_subplot(grid[-1,1:], yticklabels=[], sharex=main_ax)

    main_ax.imshow(S)
    if title != None: main_ax.set_title(title)
    # x_hist.hist(S[:,0], 40, histtype='stepfilled', orientation='vertical')
    xbardata = zoom(S.sum(axis=0), 1/rebin)
    x_hist.bar(np.linspace(0, len(S.sum(axis=0)), len(xbardata)), xbardata, width=8*rebin)
    x_hist.invert_yaxis()

    ybardata = zoom(S.sum(axis=1), 1/rebin)
    y_hist.barh(np.linspace(0, len(S.sum(axis=1)), len(ybardata)), ybardata, height=8*rebin)
    y_hist.invert_xaxis()
    return xbardata, ybardata
```

```python
for i, f in enumerate(data):
    d = io.imread(f).astype(np.float)
    xb, yb = projection_axis_bar(d, title=names[i])
```
